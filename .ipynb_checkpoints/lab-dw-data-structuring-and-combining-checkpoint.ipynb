## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here

import pandas as pd

df1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")

df1.columns = df1.columns.str.lower().str.strip()
df2.columns = df2.columns.str.lower().str.strip()
df3.columns = df3.columns.str.lower().str.strip()

df1.info
df2.info
df3.info
```

```python
# Combine all dataframes into one
df = pd.concat([df1, df2, df3], axis=0)
df
```

```python
df['state'] = df['state'].combine_first(df['st'])
df.drop('st', axis=1, inplace=True)

df
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(' ', '_')
    .str.replace('state', 'st')
)

df.columns
```

```python
df['gender'].unique()

df['gender'] = df['gender'].str.lower().replace({
    'f': 'F',
    'female': 'F',
    'femal': 'F',
    'm': 'M',
    'male': 'M'
})

state_map = {
    'az': 'Arizona',
    'cali': 'California',
    'wa': 'Washington'
}

df['education'] = df['education'].str.replace('Bachelors', 'Bachelor', case=False)

df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '')

df['vehicle_class'] = df['vehicle_class'].replace({
    'Sports Car': 'Luxury',
    'Luxury SUV': 'Luxury',
    'Luxury Car': 'Luxury'
})

df
```

```python
df.info()

df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')

print(df['customer_lifetime_value'].dtype)
```

```python
df['number_of_open_complaints'] = df['number_of_open_complaints'].astype(str).apply(
    lambda x: x.split('/')[1] if '/' in x else x)

df['number_of_open_complaints'] = pd.to_numeric(df['number_of_open_complaints'], errors='coerce')
df['number_of_open_complaints'].unique()
```

```python
df
```

```python

df['income'] = df['income'].fillna(df['income'].median())
df['monthly_premium_auto'] = df['monthly_premium_auto'].fillna(df['monthly_premium_auto'].median())
df['number_of_open_complaints'] = df['number_of_open_complaints'].fillna(df['number_of_open_complaints'].median())
df['total_claim_amount'] = df['total_claim_amount'].fillna(df['total_claim_amount'].median())
df['customer_lifetime_value'] = df['customer_lifetime_value'].fillna(df['customer_lifetime_value'].median())

df['st'] = df['st'].fillna(df['st'].mode()[0])
df['education'] = df['education'].fillna(df['education'].mode()[0])
df['policy_type'] = df['policy_type'].fillna(df['policy_type'].mode()[0])
df['vehicle_class'] = df['vehicle_class'].fillna(df['vehicle_class'].mode()[0])

df
```

```python
df = df.dropna(subset=['customer', 'gender'])

print(df.isnull().sum())
```

```python
print(df.notnull().sum())
```

```python
#Drop duplicate rows
df = df.drop_duplicates().reset_index(drop=True)

df.notnull().sum()
df.info()
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Your code goes here

import pandas as pd

customerdata = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")

customerdata
```

```python
customerdata.isnull().sum()
```

```python
#Drop duplicate rows
customerdata = customerdata.drop_duplicates().reset_index(drop=True)

customerdata.notnull().sum()
customerdata.info()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
revenue_by_channel = customerdata.pivot_table(
    index='sales_channel', 
    values='total_claim_amount', 
    aggfunc='sum'
).round(2)

# Display the result
print(revenue_by_channel)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
clv_by_gender_education = customerdata.pivot_table(
    index='gender',
    columns='education',
    values='customer_lifetime_value',
    aggfunc='mean'
).round(2)

# Display the result
print(clv_by_gender_education)
```
